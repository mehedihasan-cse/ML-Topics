```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import wordcloud
import PIL.Image
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
shape= np.array(PIL.Image.open('love.jpg'))
```

```python
text = " Preparing for an interview? Sharing 56 Interview Questions and Answers to help you get ready.Whether you're new to interviews or need a refresher, this guide covers everything from basic questions to more challenging ones. Get ready to impress and land your dream job! Also sharing some amazing learning resources by Google"
```

```python
wc = WordCloud(stopwords=STOPWORDS,
               mask= shape, background_color ='white'
               , contour_color='Black'
               , contour_width =1).generate(text)
```

```python
plt.figure(figsize=(3,3))

plt.imshow(wc)
```
